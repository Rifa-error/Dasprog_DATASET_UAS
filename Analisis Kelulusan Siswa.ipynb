```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('dataset.csv')
df['Presentase_Lulus'] = (df['Lulus'] / df['Jumlah_Siswa']) * 100

x = df['Sekolah'].unique()
y2023 = df[df['Tahun'] == 2023]['Presentase_Lulus']
y2024 = df[df['Tahun'] == 2024]['Presentase_Lulus']
pos = range(len(x))
bar_width = 0.35

plt.figure(figsize=(8,5))
plt.bar([p - bar_width/2 for p in pos], y2023, width=bar_width, label='2023')
plt.bar([p - bar_width/2 for p in pos], y2024, width=bar_width, label='2024')
plt.xticks(pos, x)
plt.ylabel('Presentase Kelulusan (%)')
plt.title('Kelulusan Siswa per Sekolah')
plt.legend()
plt.tight_layout()
plt.show()
```
